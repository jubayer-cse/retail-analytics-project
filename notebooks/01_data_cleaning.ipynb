```python
import pandas as pd

df = pd.read_csv('online_retail_II.csv', encoding='ISO-8859-1')

df = df.rename(columns={
    'Invoice': 'InvoiceNo',
    'Price': 'UnitPrice',
    'Customer ID': 'CustomerID'
})

print(df.shape)
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
print(df.duplicated().sum())
```

```python
#Reomove rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])
```

```python
print(df['CustomerID'].isnull().sum())
```

```python
#Remove cancelled orders
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
print(df['InvoiceNo'].astype(str).str.startswith('C').sum())
```

```python
#Remove non-ositive quantities and prices
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
print(((df['Quantity'] <= 0) | (df['UnitPrice'] <= 0)).sum())
```

```python
#Remove exact duplicate rows
df = df.drop_duplicates()
```

```python
print(df.duplicated().sum())
```

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df[['Quantity', 'UnitPrice', 'TotalPrice']].head()
```

```python
df['TotalPrice'].isnull().sum()
```

```python
(df['TotalPrice'] == df['Quantity'] * df['UnitPrice']).all()
```

```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['InvoiceDate'].dtype
```

```python
df['InvoiceDate'].isnull().sum()
```

```python
df['InvoiceDate'].head()
```

```python
# Sanity check
print(f"Final sape: {df.shape}")
print(df.isnull().sum())
```

```python
print("Missing CustomerID:", df['CustomerID'].isnull().sum())
print("Cancelled orders (start with C):", df['InvoiceNo'].astype(str).str.startswith('C').sum())
print("Non-positive quantity:", (df['Quantity'] <= 0).sum())
print("Non-positive price:", (df['UnitPrice'] <= 0).sum())
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.to_csv('cleaned_retail_data.csv', index =False)
```
